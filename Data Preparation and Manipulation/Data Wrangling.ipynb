```python
# Remove the column which has only one distinct element

procstat=procstat.loc[:,procstat.nunique()!=1]
procstat.nunique()
```

```python
procstat_grouped = procstat.groupby('ProducerName')
```

```python
procstat_0 = procstat_grouped.get_group(0)
procstat_1 = procstat_grouped.get_group(1)
procstat_2 = procstat_grouped.get_group(2)
procstat_3 = procstat_grouped.get_group(3)
procstat_4 = procstat_grouped.get_group(4)
procstat_5 = procstat_grouped.get_group(5)
```

```python
procstat_0.nunique()
```

```python
# Remove the column which has only one distinct element

procstat_0=procstat_0.loc[:,procstat_0.nunique()>1]
procstat_1=procstat_1.loc[:,procstat_1.nunique()>1]
procstat_2=procstat_2.loc[:,procstat_2.nunique()>1]
procstat_3=procstat_3.loc[:,procstat_3.nunique()>1]
procstat_4=procstat_4.loc[:,procstat_4.nunique()>1]
procstat_5=procstat_5.loc[:,procstat_5.nunique()>1]
procstat_0.nunique()
```
