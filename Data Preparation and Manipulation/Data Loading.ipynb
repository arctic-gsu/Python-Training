 # Loading data from different sources

## 1. Loading from CSV 

```python
import pandas as pd
```

```python
# reading data from csv
data = pd.read_csv("../data/georgia-history.csv")
data.head()
```

## 2. Loading from URL

```python
url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"

# reading data from url
data = pd.read_csv(url)
data.head()
```

## 3. Loading from Microsoft Excel

```python
# reading data from excel. Need to install package openpyxl. Can install using "pip install openpyxl"
data = pd.read_excel("../data/georgia-history.xlsx")
data.head()
```

## 4. Loading from JSON file

```python
# reading data from json
data = pd.read_json("../data/georgia-history.json")
data.head()
```

## 5. Loading from XML file

```python
# reading data from xml
data = pd.read_xml("../data/georgia-history.xml")
data.head()
```
