## Plots

```python
x = [1,2,3,4]
y = [2,4,6,8]
```

```python
plt.plot(x,y)
```

```python

# help(plt.plot)
```

```python
# Changing color of line

plt.plot(x,y,c='red')
```

```python
# Adding labels on x and y column
plt.plot(x,y,c='blue')
plt.xlabel('X label',fontsize =15, c ='red')
plt.ylabel('Y label',fontsize =15, c ='Blue')
```

```python
# Passing only one value to the plot takes it as y value and index as x value
plt.plot([1,2,3,4])
```

```python
# Adding ticks
plt.plot(x,y,c='blue')
plt.xlabel('X label',fontsize =15, c ='red')
plt.ylabel('Y label',fontsize =15, c ='Blue')
plt.tick_params(axis='x',color='red',labelsize='large',labelcolor='blue')
plt.tick_params(axis='y',color='red',labelsize='large',labelcolor='red')
```

```python
# Adding ticks
plt.plot(x,y,c='blue',label='Line')
plt.legend()
plt.xlabel('X label',fontsize =15, c ='red')
plt.ylabel('Y label',fontsize =15, c ='Blue')
plt.tick_params(axis='x',color='red',labelsize='large',labelcolor='blue')
plt.tick_params(axis='y',color='red',labelsize='large',labelcolor='red')
```

```python
# Adding title
plt.title("Line", {'fontsize':20,'color':'green'})
plt.plot(x,y,c='blue',label='Line')
plt.legend()
plt.xlabel('X label',fontsize =15, c ='red')
plt.ylabel('Y label',fontsize =15, c ='Blue')
plt.tick_params(axis='x',color='red',labelsize='large',labelcolor='blue')
plt.tick_params(axis='y',color='red',labelsize='large',labelcolor='red')
```

```python
#help(plt.title)
```

```python
# Limiting X and Y value
plt.plot(x,y,c='blue',label='Line')
plt.xlim(1,3)
plt.ylim(2,4)
```

```python
# Taking mirror image using xlim and ylim function
plt.plot(x,y,c='blue',label='Line')
plt.xlim(3,1)
plt.ylim(4,2)
```

```python
# Adding multiple lines on the same axes
import numpy as np

a =[2,4,6,8]
b = np.sqrt(a)
plt.plot(x,y,c='blue',label='Line 1')
plt.plot(a,b,c='red',label='Line 2')
plt.legend()
```

```python
# Adding style to lines
import numpy as np

a =[2,4,6,8]
b = np.sqrt(a)
plt.plot(x,y,c='blue',label='Line 1', linestyle=":")
plt.plot(a,b,c='red',label='Line 2',linestyle="--",linewidth=3)
plt.legend()
```

```python
# Focusing on data points
# Adding style to lines
import numpy as np

a =[2,4,6,8]
b = np.sqrt(a)
plt.plot(x,y,c='blue',label='Line 1', linestyle=":", marker='o')
plt.plot(a,b,c='red',label='Line 2',linestyle="--",linewidth=5, marker='d', markersize=10)
plt.legend()
```

```python
plt.scatter(a,b,marker='d',color='red')
```

## Figure

Figure is a area where plots are drawn

```python
fig = plt.figure()
ax_1 = fig.add_axes([0,0,1,1]) # left axis, bottom axis, width of  axis, height of axis
ax_2 = fig.add_axes([1.5,0,1,1])
```

```python

import numpy as np 

a = [2,4,6,8]
b =[4,16,36,64]
x = np.arange(0,150,2)
y = np.sin(np.radians(x))
fig = plt.figure()
ax_1 = fig.add_axes([0,0,1,1])
ax_2 = fig.add_axes([1.5,0,1,1])
ax_1.set_xlabel('X label',fontsize =15, c ='red')
ax_1.set_ylabel('Y label',fontsize =15, c ='Blue')
ax_1.plot(a,b)
ax_2.plot(x,y)
```

```python
# Let's make a plot inside plot

import numpy as np 

fig = plt.figure()
ax_1 = fig.add_axes([0,0,1,1])
ax_2 = fig.add_axes([0.2,0.5,0.5,0.4])

a = [2,4,6,8]
b =[4,16,36,64]
ax_1.set_xlabel('X label',fontsize =15, c ='red')
ax_1.set_ylabel('Y label',fontsize =15, c ='Blue')
ax_1.plot(a,b)

x = np.arange(0,150,2)
y = np.sin(np.radians(x))

ax_2.plot(x,y)

```

### Figure Subplot

```python
fig = plt.figure()

ax_1 = fig.add_subplot(321) # 3 by 2 grid at position1
plt.show()
```

```python
fig = plt.figure(figsize=(8,8))

ax_1 = fig.add_subplot(321)
ax_1 = fig.add_subplot(322)
ax_1 = fig.add_subplot(323)
ax_1 = fig.add_subplot(324)
ax_1 = fig.add_subplot(325)

ax_1 = fig.add_subplot(326)
plt.show()
```

### Figure Subplotgrid

```python
a = [2,4,6,8]
b =[4,16,36,64]

x = np.arange(0,150,2)
y = np.sin(np.radians(x))


ax_1 = plt.subplot2grid((2,2),(0,1)) # 2X2 grid. Create axes on 0 row and 1 column
ax_1.plot(a,b)

ax_1 = plt.subplot2grid((2,2),(1,0),colspan=2)
ax_1.plot(x,y)
```

## Working with Datasets

```python
import pandas as pd
ga_data = pd.read_csv("data/georgia-history.csv")
```

```python
ga_data.head(50)
```

```python
a = ga_data["date"]
b = ga_data["death"]

x = ga_data["date"]
y = ga_data["hospitalized"]


plt.plot(x,y,c='blue',label='Hospitalized', linestyle=":")
plt.plot(a,b,c='red',label='Death',linestyle="--")

plt.tick_params(axis='y',color='red',labelsize='large',labelcolor='red')

plt.legend()
```

## Adding shapes

```python
import matplotlib.patches as patches
fig = plt.figure(figsize=(10,10))

ax_1 = fig.add_subplot(321) 
ax_1.add_patch(patches.Rectangle((0,0),0.6,0.5,fill=False)) # left-edge, bottom-edge, width,heigh
plt.show()
```

## Annotations

```python
x = [1,2,3,4]
y = [2,4,6,8]

plt.plot(x,y)
plt.annotate('min value',xy=(1,2),xytext=(2,2),arrowprops=dict(color='b'))
```

```python
x = [1,2,3,4]
y = [2,4,6,8]

plt.plot(x,y)
plt.annotate('min value',xy=(1,2),xytext=(1,3),arrowprops=dict(color='b',alpha=0.4))
plt.plot(1,2,marker='o')
```

## Box Plot
